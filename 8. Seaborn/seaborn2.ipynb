<h1>Visualizing Statistical Relationship</h1>


<li>relplot is used to plot statistical relationships with seaborn

<LI>relplot plots a scatter plot by default

<li>A scatter plot is uesd to vizualize the relation between 2 variables <br><br>


```python
df = sns.load_dataset("tips")
df.head()
```

```python
#if the bill is more in value does that mean more tips?
```

click for documentation : <a href = "https://seaborn.pydata.org/generated/seaborn.relplot.html">relplot documentation</a> 

```python
sns.relplot(x = "total_bill", y = "tip", data = df)
#by-default a scatter plot is plotted
```

```python
#changing to line plot
sns.relplot(x = "total_bill", y = "tip", data = df, kind = "line" )
#only scatter and line plots are possible with relplot()
```

```python
#relation between bill and tip based on time of meal
sns.relplot(x = "total_bill", y = "tip", data = df, hue = "time")
```

```python
#styling the markers
sns.relplot(x = "total_bill", y = "tip", data = df, hue = "time",style = "time")
```

```python
#smoker or not?
sns.relplot(x = "total_bill", y = "tip", data = df,hue = "smoker",style = "time",col = "smoker")
```
