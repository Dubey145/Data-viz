```python
import seaborn as sns
sns.set()
```

```python
df = sns.load_dataset("flights")
df.head()
```

```python
sns.relplot(x = "year", y = "passengers",data = df, kind = "line")
```

```python
sns.relplot(x = "year", y = "passengers",data = df, kind = "scatter")
```

```python
sns.lineplot(x = "year", y = "passengers",data = df)
```

```python
 sns.lineplot(x = "year", y = "passengers",data = df, hue = "month")
```

```python
#plotting for each month
sns.relplot(x = "year", y = "passengers",data = df, hue = "month",col = "month")
```
