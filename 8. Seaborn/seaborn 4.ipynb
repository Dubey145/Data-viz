# representing categorical data
- a categorical column is a column on which we cannot perform mathematical operations 
eg months
- catplot( ) is used from seaborn
- 

# categorical scatter plot

the caregorical data usually goes on the x axis and the y axis has the numerical value

the strip plot is the default plot 

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.catplot(x = "sex",y = "total_bill", data = df)
#the default strip plot
# a strip plot  It is basically a scatter plot that differentiates different categories.
#did not execute sns.set() notice the difference 
```

```python
sns.set()
sns.relplot(x = "sex",y = "total_bill", data = df)
```

```python
#using jitter to control randomness of the points
#default jitter value is 1
sns.catplot(x = "sex",y = "total_bill", data = df,jitter = 0.25)
```

```python
sns.catplot(x = "sex",y = "total_bill", data = df,jitter = 0.80)
```

```python
#swarmplot
#does not allow overlapping of markers
sns.swarmplot(x = "sex",y = "total_bill",data = df)
```

```python
#using hue
sns.swarmplot(x = "sex",y = "total_bill",data = df, hue = "smoker")
```

# categorical Distribution plot

we use distribution plots to understand how values are distributed in the dataset 

**boxplot( )**


<pre>
lower quartile is the 25th percentile 
median is the 50th percentile 
upper quartile is the 75th percentile 
</pre>
**note :** any value outside the min or max marker will be considered as an outlier 

```python
sns.boxplot(x = "total_bill", y  = "sex",data = df)
```

```python
sns.boxplot(x = "sex", y  = "total_bill",data = df)
```

```python
sns.boxplot(x = "sex", y  = "total_bill",data = df,hue = "smoker")
```

- a boxen plot is similat to boxplot but it has different quartile values 

```python
sns.boxenplot(x = "sex", y  = "total_bill",data = df,hue = "smoker")
```
